```python
from matplotlib import pyplot as plt
from astroML.datasets import fetch_sdss_spectrum

#------------------------------------------------------------
# Fetch single spectrum
plate = 2215
mjd = 53793
fiber = 296

spec = fetch_sdss_spectrum(plate, mjd, fiber)

#------------------------------------------------------------
# Plot the resulting spectrum
ax = plt.axes()
ax.plot(spec.wavelength(), spec.spectrum, label='spectrum') # 'k' optional for gray
ax.plot(spec.wavelength(), spec.error, '-', color='red', label='error')

ax.legend(loc=1)

ax.set_title('Plate = %(plate)i, MJD = %(mjd)i, Fiber = %(fiber)i' % locals())

ax.text(0.05, 0.95, 'z = %.2f' % spec.z, size=16,
        ha='left', va='top', transform=ax.transAxes)

ax.set_xlabel(r'$\lambda (\AA)$')
ax.set_ylabel('Flux')
ax.set_ylim(-10, 300)

plt.grid()
plt.savefig('2215_1.png')
plt.show()
```

Details: The title indicates the observation specifics: Plate = 2215, MJD = 53793, Fiber = 296, and z = 0.03 (likely the redshift value).
