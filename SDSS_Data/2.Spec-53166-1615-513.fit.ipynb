Every Header Data Unit (HDU) normally has two components: header and data. In astropy these two components are accessed through the two attributes of the HDU, hdu.header and hdu.data.

The terms you mentioned are common in FITS (Flexible Image Transport System) data files, which are widely used in astronomy for storing, transmitting, and analyzing scientific data. Hereâ€™s a brief explanation of each term:

No.: This likely refers to the number or index of the data entry.
Name: The name or identifier of the data entry.
Ver: The version of the data or software used.
Type: The type of data or object (e.g., image, table).
Cards: This refers to the number of header cards in the FITS file. Each card contains metadata about the data.
Dimensions: The dimensions of the data array (e.g., 2D image, 3D data cube).
Format: The format of the data (e.g., integer, floating-point).
